##### Main

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
from scipy.special import factorial
from utils import g
```

```python
np.random.seed(0)

# Dimension of the problem
K = 3

# Order we will go up to
N = 10

# Random covariance and its Cholesky decomp.
L = 0.1 * np.tril(0.25 * np.random.randn(K, K) + np.eye(K))
Sig = L @ L.T

# Ones vector & matrix
j = np.ones((K, 1))
J = np.ones((K, K))
```

Analytic expression for $\mathbf{Q}_n$:

```python
def Q(n):
    return (n + 1) * g(n + 1) / K ** n * np.linalg.matrix_power(Sig @ J, (n + 1) // 2)


Q_analytic = np.array([Q(n) for n in np.arange(N)])
```

Numerical estimate of $\mathbf{Q}_n$:

```python
# Number of samples we'll draw
M = 1000000

# Compute samples; 
# the mean over all samples is the value of Q
q = np.empty((N, K, K, M))
for m in tqdm(range(M)):
    u = np.random.randn(K, 1)
    for n in range(N):
        q[n, :, :, m] = (n + 1) / K ** n * ((j.T @ L @ u) ** n * (L @ u @ j.T))
        
Q_numerical = np.mean(q, axis=-1)
```

```python
fig, ax = plt.subplots(K, K, figsize=(12, 10))
n = np.arange(N)
for k1 in range(K):
    for k2 in range(K):
        ax[k1, k2].plot(n[1::2], Q_numerical[:, k1, k2][1::2], "o", ms=8, color="C1")
        ax[k1, k2].plot(
            n[::2], Q_numerical[:, k1, k2][::2], "o", ms=8, color="C1", alpha=0.1
        )
        ax[k1,k2].plot(n, Q_analytic[:, k1, k2], 'o', ms=5, color="C0")
        ax[k1, k2].set_yscale("log")
```
